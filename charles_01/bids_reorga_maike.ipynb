rename `"TaskName": "fmri"` & add `"task: "chase"`
in task.json files

```python
import os
import os.path as op
import re
from pathlib import Path
#import shutil

import glob
import json

bids_folder = '/mnt_01/ds-asd'
subList = [id[4:] for id in os.listdir(bids_folder) if re.match(r'sub-\d{2}', id)]
```

```python
# subList for submitting fmriprep sbatch script
import numpy as np

subList = [int(id[4:]) for id in os.listdir(bids_folder) if re.match(r'sub-\d{2}', id)]
print(','.join(str(sub) for sub in np.sort(subList)))
```

```python
sub = '01'
fn_task_json = op.join(bids_folder, f'sub-{sub}/ses-1/func/sub-{sub}_ses-1_task-chase_run-1_bold.json')
```

```python
pati = op.join(bids_folder, f'sub-{sub}/ses-1/func')
fns = glob.glob(op.join(pati, f'sub-{sub}_ses-1_task-chase_run-*_bold.json'))
print(fns)
```

```python
for fn_task_json in fns:
    with open(fn_task_json, 'r') as f: # Load the JSON file
        data = json.load(f)

    data["TaskName"] = "Chase task" # Replace a value (this key existed before)
    data["task"] = "chase" #  add a new key (this key did not exist before)

    with open(fn_task_json, 'w') as f: # Save the changes back to the file
        json.dump(data, f, indent=4)
```

```python
for sub in subList:
    pati = op.join(bids_folder, f'sub-{sub}/ses-1/func')
    fns = glob.glob(op.join(pati, f'sub-{sub}_ses-1_task-chase_run-*_bold.json'))
    for fn_task_json in fns:
        with open(fn_task_json, 'r') as f: # Load the JSON file
            data = json.load(f)

            data["TaskName"] = "Chase task" # Replace a value (this key existed before)
            data["task"] = "chase" #  add a new key (this key did not exist before)
            
        with open(fn_task_json, 'w') as f: # Save the changes back to the file
            json.dump(data, f, indent=4)
    print(f'Finished sub-{sub}')
```
