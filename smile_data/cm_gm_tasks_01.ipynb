```python
import os
import os.path as op
import pandas as pd
import numpy as np
import nibabel as nib
from scipy.stats import normaltest, ttest_ind, mannwhitneyu
import matplotlib.pyplot as plt
import seaborn as sns

datapool_root_folder = '/mnt_AdaBD_largefiles/Data/DNumRisk_Data' #'/mnt_AdaBD_largefiles/Data/SMILE_Data/DNumRisk'
dataset = 'smile'
bids_folder = op.join(datapool_root_folder, f'ds-{dataset}')

subList = [int(d[4:]) for d in os.listdir(bids_folder) if d.startswith('sub-')]
```

```python
sub = 102
cm_folder = op.join(bids_folder,'derivatives','correlation_matrices')
cm_rest = np.load(op.join(cm_folder, f'sub-{sub}',  f'sub-{sub}_ses-1-2_task-rest_funcCM.npy'))
cm_magjudge = np.load(op.join(cm_folder, f'sub-{sub}',  f'sub-{sub}_ses-1-2_task-magjudge_funcCM.npy'))
cm_placevalue = np.load(op.join(cm_folder, f'sub-{sub}',  f'sub-{sub}_ses-1-2_task-placevalue_funcCM.npy'))
```

```python
fig, ax = plt.subplots(figsize=(3,3))
sns.histplot(cm_magjudge.flatten(), bins=100, color='red', label='magjudge',ax=ax, alpha=0.4)
sns.histplot(cm_rest.flatten(), bins=100, color='green', label='rest',ax=ax, alpha=0.4)
sns.histplot(cm_placevalue.flatten(), bins=100, color='blue', label='placevalue',ax=ax, alpha=0.4)
ax.legend(loc='upper right', bbox_to_anchor=(1,1))
sns.despine()
ax.set(title=f'Subject {sub} Correlation Matrix Values')
plt.tight_layout()
plt.show()
```
