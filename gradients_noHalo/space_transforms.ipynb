```python
# construct Margulies reference gradient in fsaverage5 space in same shape
from neuromaps.datasets import fetch_annotation
margulies2016 = fetch_annotation(source='margulies2016')

grads = [None]*10
for n_grad in range(1,11):
    grads_hemis = [None]*2
    for n_hemi, hemi in enumerate(['L','R']):
        grad_fslr_fn = margulies2016['margulies2016',f'fcgradient{n_grad:02d}','fsLR','32k'][n_hemi]
        gii_im_fsav = transforms.fslr_to_fsaverage(grad_fslr_fn, '10k',hemi=hemi)
        grads_hemis[n_hemi] = gii_im_fsav[0].agg_data()
    grads[n_grad-1] = np.concatenate(grads_hemis, axis=0)

target_fn = op.join(bids_folder,'derivatives', 'gradients','reference_gradients_margulies16_space-fsaverage5_N-10.npy')
np.save(target_fn, np.array(grads))
```
