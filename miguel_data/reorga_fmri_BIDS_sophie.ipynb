```python
import os
import shutil
import re

# Source and destination directories
src_root = "/mnt_04/ds-numrisk/derivatives/fmriprep"
dst_root = "/mnt_AdaBD_largefiles/Data/SMILE_Data/DNumRisk/ds-numrisk/derivatives/fmriprep"

# Session
ses = "1"

# Patterns of files to keep
keep_patterns = [
    "space-fsaverage5",
    "space-T1w_desc-brain_mask",
    "space-T1w_desc-preproc_bold"
]

# Mapping for renaming tasks
task_rename = {
    "numrisk": "magjudge"
}

# Regex to detect sub number in filenames
sub_pattern = re.compile(r"sub-(\d+)(?=_)")

# Function to zero-pad subject folder names
def format_sub(sub_name):
    match = re.match(r"sub-(\d+)", sub_name)
    if match:
        sub_num = int(match.group(1))
        return f"sub-{sub_num:02d}"
    return sub_name


# Main copying and renaming loop
for sub_name in os.listdir(src_root):
    if not sub_name.startswith("sub-"):
        continue

    src_func_dir = os.path.join(src_root, sub_name, "func")
    if not os.path.exists(src_func_dir):
        continue

    # Zero-pad subject folder
    new_sub_name = format_sub(sub_name)

    # Correct folder structure: sub-XX/ses-1/func
    dst_func_dir = os.path.join(dst_root, new_sub_name, f"ses-{ses}", "func")
    os.makedirs(dst_func_dir, exist_ok=True)

    # Iterate over files in func
    for file_name in os.listdir(src_func_dir):
        # Keep only files matching certain patterns
        if not any(pat in file_name for pat in keep_patterns):
            continue

        new_file_name = file_name

        # Dynamically rename task
        for old_task, new_task in task_rename.items():
            if f"_task-{old_task}" in new_file_name:
                new_file_name = new_file_name.replace(f"_task-{old_task}", f"_task-{new_task}")

        # Remove acq-* substring (e.g., _acq-9)
        new_file_name = re.sub(r"_acq-\d+", "", new_file_name)

        # Zero-pad the subject number inside the filename (sub-1 → sub-01)
        new_file_name = sub_pattern.sub(lambda m: f"sub-{int(m.group(1)):02d}", new_file_name)

        # Add session to filename if not already there (e.g., sub-01 → sub-01_ses-1)
        match_sub = re.search(r"(sub-\d{2})", new_file_name)
        if match_sub:
            sub_id = match_sub.group(1)
            if f"{sub_id}_ses-{ses}" not in new_file_name:
                new_file_name = new_file_name.replace(sub_id, f"{sub_id}_ses-{ses}")

        # Copy file
        src_file = os.path.join(src_func_dir, file_name)
        dst_file = os.path.join(dst_func_dir, new_file_name)

        if os.path.exists(dst_file):
            print(f"Skipped (exists): {dst_file}")
            continue

        shutil.copy2(src_file, dst_file)
        print(f"Copied {src_file} → {dst_file}")

print(" Copying and renaming completed.")
```

```python
# reorganization of event files of numrisk so they match event files of dnumrisk

import os.path as op
import pandas as pd
import os
import warnings

warnings.filterwarnings("ignore")

input_dir = '/mnt_04/ds-numrisk'
output_dir = '/mnt_AdaBD_largefiles/Data/SMILE_Data/DNumRisk/ds-numrisk/'

ses = '1'
runs = ['1', '2', '3', '4', '5', '6']
#runs = ['1']

subList = sorted([f.split('-')[1] for f in os.listdir(input_dir)
                  if f.startswith("sub-")], key=lambda x: int(x))

subList = subList[1:] # only if input folder contains sub-0 as a pilot sub

#subList = ['01']

#final_cols = ['trial_nr', 'onset', 'trial_type', 'n1', 'n2', 'choice']

for sub in subList:

    counter = 0

    # zero-pad for output folder/file
    sub_padded = str(int(sub)).zfill(2)  # e.g., '1' -> '01'

    for run in runs:
        sub_dir = op.join(input_dir, f'sub-{sub}', 'func', f'sub-{sub}_task-numrisk_run-{run}_events.tsv') #f'sub-{sub}_ses-{ses}_task-magjudge_run-{run}_events.tsv'
        event_file = pd.read_csv(sub_dir, sep="\t")

        #if list(event_file.columns) == final_cols:
            #print(f"Skipping {sub_dir} (already formatted)")
            #continue


        event_file['n1'] = None
        event_file['n2'] = None
        event_file['choice'] = None
        event_file['trial_nr'] = None

        event_file = event_file.drop(columns=['duration', 'response_time'])

        for i, entry in enumerate(event_file['trial_type']):
            if "stim1-" in entry:
                counter+=1

            event_file['trial_nr'][i] = counter

            if entry == 'left':
                event_file['choice'][i] = 2
                event_file['trial_type'][i] = 'choice'

            elif entry == 'right':
                event_file['choice'][i] = 1
                event_file['trial_type'][i] = 'choice'

            if 'stim1-'in entry:
                event_file['n1'][i] = entry[6:]
                event_file['n1'][i + 1] = entry[6:]
                event_file['trial_type'][i] = 'stimulus 1'

            elif 'stim2-'in entry:
                event_file['n2'][i] = entry[6:]
                event_file['n2'][i - 1] = entry[6:]
                event_file['trial_type'][i] = 'stimulus 2'
            
        event_file = event_file.loc[:, ['trial_nr', 'onset', 'trial_type', 'n1', 'n2', 'choice']]

        output_subdir = op.join(output_dir, f'sub-{sub_padded}', 'ses-1', 'func')
        os.makedirs(output_subdir, exist_ok=True)

        save_dir = op.join(output_subdir, f'sub-{sub_padded}_ses-{ses}_task-magjudge_run-{run}_events.tsv')
        event_file.to_csv(save_dir, sep="\t", index=False)




    print(f'sub-{sub_padded} done')

print(event_file)



```
